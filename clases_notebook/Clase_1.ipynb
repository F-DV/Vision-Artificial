#2. Utilizando condicionales

```python
#Computando horas de trabajo (Pilas con el tipo de datos)
#Ojo: hasta 40 hooras valen loo mismo, de ahí en adelante tienen el 50% de recargo
print('Hello, welcome to the weekly salary calculator')
Hours=input('How many hours did you work this week?\n')
try:
    Hours=int(Hours)
except:
    Hours=input('You Hour data must be a number, enter it again\n')
    Hours=int(Hours)

Price=input('How much is per hour?\n')
try:
    Price=float(Price)
except:
    Price=input('You Hour value must be a number, enter it again\n')
    Price=float(Price)

if Hours>40:
    Pay=(40*Price)+((Hours-40)*Price*1.5)
else:
    Pay=(Hours*Price)

print('Your pay will be',Pay,'pesos!')
```

```python
n = 10
while True:
  print(n)
  if n == 0:
    break
  n=n-1
    
print('Done!')
```

```python
#Importando librerías
import cv2 # OpenCV para computer vision
import numpy as np # Para cálculo de matrices
import matplotlib.pyplot as plt #Para graficar

#Leyendo una imagen en grises
path = './cameraman.tif'#Indica el direcorio de la imagen
Img = cv2.imread(path,0) #Lee la imagen en escala de grises
plt.imshow(Img,cmap='gray', vmin=0, vmax=255) #Grafica la imagen en campo de grises
plt.show
```
