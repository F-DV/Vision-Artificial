Importar librerias

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Leer image

```python
IMG = cv2.imread('./Lena.png')
IMG = IMG[:,:,[2,1,0]]
```

```python
plt.imshow(IMG)
plt.show()
```

Recorte

```python
RECORTE= IMG[200:319,200:309,:]

#Visualizando imagen recortada
print('Recorte')
plt.imshow(RECORTE,vmax=255, vmin=0)
plt.show()
```

Subir intensidad

```python
[ROWS,COLS,CH] = IMG.shape
print("({}, {}, {})".format(ROWS,COLS,CH))
```

```python
RES=np.zeros([ROWS,COLS,CH])

for i in range(0,COLS):
  for j in range(0,ROWS):
    for k in range(0,CH):
      RES[j,i,k] = IMG[j,i,k]+150
      if RES[j,i,k]>=255:
        RES[j,i,k]=255
      else:
        RES[j,i,k]=RES[j,i,k]
```

```python
plt.imshow(RES.astype('uint8'),vmin=0,vmax=255)
plt.show()
```

Bajar intensidad

```python
RES=np.zeros([ROWS,COLS,CH])

for i in range(0,COLS):
  for j in range(0,ROWS):
    for k in range(0,CH):
      RES[j,i,k] = IMG[j,i,k]-50
      if RES[j,i,k]<=0:
        RES[j,i,k]=0
      else:
        RES[j,i,k]=RES[j,i,k]
```

```python
plt.imshow(RES.astype('uint8'),vmin=0,vmax=255)
plt.show()
```
